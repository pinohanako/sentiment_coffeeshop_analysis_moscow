**Диаграмма распределения количества отзывов с течением времени**

На оси X откладываются даты, на оси Y - плотность (или количество) отзывов в диапазоне. Плотность на графике выражена в диапазоне от 0 до 0.0008, поскольку это является результатом нормализации данных для визуализации.

Различные кофейни представлены разными цветами, что позволяет сравнивать количество отзывов для разных заведений и делать определенные выводы об их популярности в разный период

*Нормализация заключается в том, что значения плотности масштабируются в определенный диапазон*

**Распределение рейтингов отзывов по датам для разных кофеен**

Визуализация плотности (KDE) позволяет выявить тенденции в изменении оценок клиентов с течением времени. Глубина цвета на графике соответствует плотности (частоте) оценок, что позволяет визуально оценить частоту встречаемости оценок пользователей для каждой кофейни.

```python
all_preprocessed_reviews['Дата'] = pd.to_datetime(all_preprocessed_reviews['Дата'])
all_preprocessed_reviews['Рейтинг'] = all_preprocessed_reviews['Рейтинг'].astype(int)
ax = sns.kdeplot(
   data=all_preprocessed_reviews, x="Дата", y="Рейтинг", hue="Кофейня",
   fill=True, common_norm=False, palette="crest",
   alpha=.5,
)
sns.move_legend(ax, "lower left")
plt.title("Распределение рейтингов по датам для разных кофеен")
```

**Выводы**

**Surf Coffee** демонстрирует стабильно высокие рейтинги на протяжении всего периода, что указывает на его постоянное высокое качество обслуживания и продуктов.

**Star Hit Cafe** характеризуется более разнообразным распределением рейтингов, что может указывать на определенные колебания в качестве или на наличие более широкого спектра мнений среди посетителей.

**Prime** характеризуется стабильным, но умеренным уровнем рейтингов, что указывает на его постоянное, но не самое выдающееся качество среди конкурентов.
